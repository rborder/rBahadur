```r
mbreaks <- function(xx,base=2, depth=3) 
    sort(unique(as.vector(sapply(xx, function(x) log(seq(base^x,base^(x+1),length.out=subdivisions),base = base)))))
```

```r
mbreak <- function(x,base=2, depth=3, j=1) {
        if (j==1) x <- c(x,x+1)
        nn <- length(x)
        tmp <- sort(unique(c(x, log(seq(base^x[nn-1],base^x[nn],length.out=3), base=base))))
    if (j<depth) {
        return(sort(unique(c(x, mbreak(tmp, base=base, depth=depth, j+1)))))
        }
    return(tmp)
    }
mbreak <- function(x,base=2, depth=3, j=1) {
        if (j==1) x <- c(x,x+1)
        nn <- length(x)
        tmp <- sort(unique(c(x, seq(x[nn-1],x[nn],length.out=3))))
    if (j<depth) {
        return(sort(unique(c(x, mbreak(tmp, base=base, depth=depth, j+1)))))
        }
    return(tmp)
    }
mbreaks <- function(xx, BASE=2, DEPTH=4) sort(unique(as.vector(sapply(xx, mbreak, base=BASE, depth=DEPTH))))
```

```r
simres <- rbind.data.frame(simres2, big_runsx)
# rownames(simres) <- NULL
# simres



# simres$run <- 1:nrow(simres)
library(reshape2)

pdat <- aggregate(timing ~ m+method, simres, function(x) mean(log10(x)))
pdat$sd <- cbind.data.frame(aggregate(timing ~ m+method, simres, function(x) sd(log10(x))))[[3]]

library(ggplot2)
names(simres)
library(splines)
ggplot(pdat, aes(log2(m), timing, ymax= timing +sd, ymin = timing-sd, color=method)) + geom_pointrange() + geom_line()

```
